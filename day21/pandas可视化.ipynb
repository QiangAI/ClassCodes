- 把数据发en线图
    - 5 个值
        - 上边缘（异常值外的最大值）
        - 下边缘（异常值外的最小值）
        - 上四分位数：Q3：上四分位：判定是否在位置上，在就直接取这个值：Q3，不在：线性插值计算上四分位数
        - 中位数：Q2  ：中位
        - 下四分位数：Q1：下四分位
        
        - 上限：Q3 + 1.5*（Q3-Q1）
        - 下限：Q1  - 1.5*（Q3-Q1）
        - 判断异常值->取上边缘，取下边缘。
    - 把数据（非异常值）分成4分0-25%-50%-75%-100%
