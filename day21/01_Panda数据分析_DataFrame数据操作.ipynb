5. 使用dtype指定数据类型

````python
import pandas as pd
import numpy as np

data = [
    [1,2,3,4,5],
    [4,5,6,7,8]
]

df = pd.DataFrame(
    data=data, 
    columns=['列1', '列2', '列3', '列4', '列5'], 
    index=['行1','行2'], 
    dtype=np.float )
print(df)
````

6. copy参数的使用

    - copy参数只对data为DataFrame与2维ndarray有效。

````python
import pandas as pd
import numpy as np

data1 = [
    [1,2,3,4,5],
    [4,5,6,7,8]
]

df1 = pd.DataFrame(
    data=data1, 
    columns=['列1', '列2', '列3', '列4', '列5'], 
    index=['行1','行2'], 
    copy=False
)
data1[1][1]=88
print('改变data（类型list）')
print(df1)
# ------------------------
data2 = np.array([
    [1,2,3,4,5],
    [4,5,6,7,8]
])

df2 = pd.DataFrame(
    data=data2, 
    columns=['列1', '列2', '列3', '列4', '列5'], 
    index=['行1','行2'], 
    copy=False
)
data2[1][1]=88   #  通过ndarray改变DataFrame
print('-----------------')
print('改变data（类型ndarrray）')
print(df2)

print('-----------------')
df2['列1'].iloc[0]=99   #  通过DataFrame改变ndarray
print('改变DataFrame，影响原来的input', data2)
````

### where按条件访问

- where 可以使用条件表达式查询

```python

DataFrame.where(
    cond,     # bool值数组
    other=nan,    # 不满足条件的返回数据
    inplace=False,  # 是否对数据直接操作
    axis=None, 
    level=None, 
    errors='raise', 
    try_cast=False, 
    raise_on_error=None)

```

````python
import pandas as pd
import numpy as np

alg = pd.read_csv('alg.csv')

re = alg.where(alg['job_listed']=='C轮', other=0)  # 返回C轮的数据，其他数据为NaN。 cond的形状必须与self保持一致
print(re[1:3])
````

### 使用mask访问数据集

- 与where刚好条件相反，where是显示满足条件的行，mask是隐藏满足条件的行。
```python
DataFrame.mask(
    cond,     # 隐藏条件
    other=nan,     # 显示NaN
    inplace=False, 
    axis=None, 
    level=None, 
    errors='raise', 
    try_cast=False, 
    raise_on_error=None)
```

````python
import pandas as pd
import numpy as np

alg = pd.read_csv('alg.csv')

re = alg.mask(alg['job_listed']=='C轮', other=0)  # 返回C轮的数据，其他数据为NaN。 cond的形状必须与self保持一致
print(re[1:3])

````

1. apply按照列处理
     - axis = 0

````python
import pandas as pd
import numpy as np

data = pd.DataFrame({'列1': [1, 2], '列2': [3, 4]})

def data_handle(rc):
    print(rc)   # 列：rc是一列数据
    return rc
    
re = data.apply(data_handle)
print(re)
````

2. apply按照行处理
    - axis = 1

````python
import pandas as pd
import numpy as np

data = pd.DataFrame({'列1': [1, 2], '列2': [3, 4]})

def data_handle(rc):
#     print(rc)     # rc 是一行数据
    return rc.sum()

re = data.apply(data_handle,axis=1)   # 每行求和
print(re)
````

3. applymap的应用
    - 应用到每个数据元素

````python
import pandas as pd
import numpy as np

data = pd.DataFrame({'列1': [8, 7], '列2': [3, 4]})

def data_handle(rc):   # 第一列被调用了两次
    print(rc)      # rc是每个元素的值
    print(type(rc))
    return rc**2

re = data.applymap(data_handle)   # 每行求和
print(re)
````

4. pipe的应用
    - 对数据集连续处理

````python
import pandas as pd
import numpy as np

data = pd.DataFrame({'列1': [8, 7], '列2': [3, 4]})

def data_handle(rc, p):   
    print(rc)   # 需要处理的数据集，rc是data的所有数据DataFrame
    print(p)   # 传递的参数
    print(type(rc))
    return rc**2

re = data.pipe(data_handle, '参数')   # 每行求和
print(re)
````

### 滑动窗口


- 滑动窗口是另外一种指定统计范围的方式。

```python

    DataFrame.rolling(
        window, 
        min_periods=None, 
        center=False, 
        win_type=None, 
        on=None, 
        axis=0, 
        closed=None)

```



#### 参数说明

- window参数
    - 整数类型，表示窗口大小，指定统计数据的行数或者列数。

- center
    - 一处理标签作为中心

- win_type参数
    - 字符串类型，由预先定义的字符串定义窗口类型，用来控制加权方式。
        - boxcar
        - triang
        - blackman
        - hamming
        - bartlett
        - parzen
        - bohman
        - blackmanharris
        - nuttall
        - barthann
        - kaiser (needs beta)
        - gaussian (needs std)
        - general_gaussian (needs power, width)
        - slepian (needs width).
- on
    - 指定不滑动的列

- axis
    - 指定滑动窗口的维度（指定0或者1效果都一样）

### DataFrame保存例子

1. 默认保存
     - 自动保存索引

````python
import pandas as pd
import numpy as np


alg = pd.read_csv('alg.csv', usecols=['job_company', 'job_listed', 'job_salary'])
java = pd.read_csv('java.csv', usecols=['job_company', 'job_listed', 'job_salary'])   # 列名一致


job = alg.append(java)

job.to_csv('job.csv', encoding='gbk')    # 由于excel等工具默认采用gbk打开汉字excel，可以保存为gbk。
````

2. 不保存索引与列名
    - 设置index=False与header=False关闭对索引与列名的保存
    

````python
import pandas as pd
import numpy as np


alg = pd.read_csv('alg.csv', usecols=['job_company', 'job_listed', 'job_salary'])
java = pd.read_csv('java.csv', usecols=['job_company', 'job_listed', 'job_salary'])   # 列名一致


job = alg.append(java)

job.to_csv('job.csv', encoding='gbk', index=False, header=False)    # 由于excel等工具默认采用gbk打开汉字excel，可以保存为gbk。

````

3. 指定保存的列

````python
import pandas as pd
import numpy as np


alg = pd.read_csv('alg.csv', usecols=['job_company', 'job_listed', 'job_salary'])
java = pd.read_csv('java.csv', usecols=['job_company', 'job_listed', 'job_salary'])   # 列名一致


job = alg.append(java)

job.to_csv('job.csv', encoding='gbk', columns=['job_company'])    # 由于excel等工具默认采用gbk打开汉字excel，可以保存为gbk。
````
