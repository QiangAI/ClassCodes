# 降维算法PCA

- 降维：
    - 相关特征去掉（特征融合）
    - 不重要特征去掉（特征选择）

- 方法：
    - 直接对数据奇异值分解(SVD)
    - PCA（基于相关性的奇异值分解）
    - FA(factor Analysis)
    - IPA(Indepedent  Component Analysis)

## sklearn的降维的实现

- iris鸢尾花（4维特征->3 ,2 , 1）

- 降维以后的数据的可分性的变化。

- PCA降维是数据预处理的一种重要的手段。

````python
"""
1. PCA类的定义说明
```python

    class sklearn.decomposition.PCA(
        n_components=None,     # 指定保留的特征数量（降维以后的维数）
        copy=True,                    # 训练过程中是否影响原始样本数据，也影响到最后降维数据的返回方式
        whiten=False,                # 是否对降维后的数据提供白化处理。（白化就是每一维的特征做一个标准差归一化处理，除以一个标准偏差） 
        svd_solver=’auto’,          #  SVD分解的计算方法
        tol=0.0,                         # 计算奇异值的公差
        iterated_power=’auto’,    # 采用svd_solver采用randomized方式的迭代次数。
        random_state=None)      # 同来创建随机数的实例，比如随机种

```
"""
% matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mlp
import numpy as np
import sklearn.datasets as ds
from sklearn.decomposition import PCA

# 1. 构建PCA对象（参数）
pca = PCA(n_components=2)

# 2. 使用数据训练（找到主要的特征值与特征向量）
# 加载数据
data, target = ds.load_iris(return_X_y=True)
# data = data[50:150, :]
# target = target[0:100]
# 降维计算（计算协方差矩阵，奇异值分解，特征值与特征向量，根据特征值大的特征向量形成新的特征空间）
pca.fit(data)   #  产生几个数据（有助于理解PCA原理）：选择以后特征向量，特征值
# print(pca.components_)    # 主成分（选择后的特征向量）
# print(pca.singular_values_)    # 主要的特征值（选择以后）
# print(pca.n_components_)   # 维数 
# 3. 对已知的数据降维

de_data = pca.transform(data)
# print(de_data)

# 降维的效果，对原来的数据可视化做对比
figure = plt.figure('PCA', figsize=(12, 18))
# 产生四个坐标(PCA降维，4个特征1，3|  2，4 |  1，4 
ax1  = figure.add_subplot(321)
ax1.set_title('PCA')
ax2 = figure.add_subplot(322)
ax2.set_title('1，3特征')
ax3 = figure.add_subplot(323)
ax3.set_title('2，4特征')
ax4 = figure.add_subplot(324)
ax4.set_title('1，4特征')
ax5 = figure.add_subplot(325)
ax5.set_title('2，3特征')

colors = mlp.colors.ListedColormap([(1,0,0), (0,1,0), (0,0,1)])
ax1.scatter(de_data[:, 0], de_data[:, 1] ,s=9, c=target,  cmap=colors)

ax2.scatter(data[:, 0], data[:, 2] ,s=9, c=target,  cmap=colors)

ax3.scatter(data[:, 1], data[:, 3] ,s=9, c=target,  cmap=colors)

ax4.scatter(data[:, 0], data[:, 3] ,s=9, c=target,  cmap=colors)

ax5.scatter(data[:, 1], data[:, 2] ,s=9, c=target,  cmap=colors)

plt.show()
````

- 每个特征变量的值，到均值的距离和称为方差（分散程度）

- $V(X)= \dfrac{1}{n} \sum \limits _{i=1}^n (x_i-\bar{X})^2$
-  $V(X, X)=\dfrac{1}{n} \sum \limits _{i=1}^n (x_i-\bar{X}) (x_i-\bar{X})$

### 协方差的概念

- 注意：
    - 特征变量

- 方差就是特殊协方差：
    - 如果特征变量一样，就是方差
    - 如果特征变量是不同，就是协方差
-  $V(X, X)=\dfrac{1}{n} \sum \limits _{i=1}^n (x_i-\bar{X}) (x_i-\bar{X})$
-  $cov(X, Y)=\dfrac{1}{n-1} \sum \limits _{i=1}^n (x_i-\bar{X}) (y_i-\bar{Y})$   : 协方差：特征之间的融合的分散程度的评估: 无偏

### 完整的PCA算法实现
    - 对特征求均值
    - 中心化
    - 计算方差

````python
% matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mlp
import numpy as np
import sklearn.datasets  as ds
import numpy.linalg as la


# 备注：降维到2
# 1. 加载数据
data, target = ds.load_iris(return_X_y=True)
data = np.mat(data)    # 矩阵与darrray的区别=：矩阵任何操作保持形状不改变，但ndarray取行，就从2维变成1维
# 2. 计算均值
data_mean = np.mean(data, axis=0)
# 3. 计算协方差矩阵
data_center = data - data_mean
data_cov = (data_center.T @ data_center )/(len(data) - 1)
# data_cov = np.matmul(data_center.T, data_center )/(len(data) - 1)
# data_cov = np.dot(data_center.T, data_center )/(len(data) - 1)    # 向量

# 4. 求奇异值分解
u, s, v = la.svd(data_cov)

# 5. 选择特征向量
e = u[:, 0:2]    # 选择两个特征
# 6. 对加载数据降维
de_data = data  @  e
de_data = np.mean(de_data) - de_data    # 中心化还原
# 7. 可视化(两个坐标系)
figure = plt.figure('我的PCA', figsize=(18, 6))

ax_my = figure.add_subplot(121, title='我的PCA算法')

colors = mlp.colors.ListedColormap([(0,1,0),(0,0,1),(1,0,0)])
ax_my.scatter(de_data[:, 0].flat, de_data[:, 1].flat , c=target, cmap=colors, s=3**2, marker='.')



ax_sk = figure.add_subplot(122, title='sklearn的PCA算法')

from sklearn.decomposition import PCA

pca = PCA(2)
pca.fit(data)

sk_data = pca.transform(data)

ax_sk.scatter(sk_data[:, 0].flat, sk_data[:, 1].flat , c=target, cmap=colors, s=3**2, marker='.')
# 8. 使用svm进行分类

# 9. 可视化分类效果

plt.show()
````

````python
% matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mlp
import numpy as np
import sklearn.datasets  as ds
import numpy.linalg as la
import sklearn as sk


# 备注：降维到2
# 1. 加载数据
data, target = ds.load_iris(return_X_y=True)
data = data[50:150,:]
target = target[0:100]

data = np.mat(data)    # 矩阵与darrray的区别=：矩阵任何操作保持形状不改变，但ndarray取行，就从2维变成1维
# 2. 计算均值
data_mean = np.mean(data, axis=0)
# 3. 计算协方差矩阵
data_center = data - data_mean
data_cov = (data_center.T @ data_center )/(len(data) - 1)
# data_cov = np.matmul(data_center.T, data_center )/(len(data) - 1)
# data_cov = np.dot(data_center.T, data_center )/(len(data) - 1)    # 向量

# 4. 求奇异值分解
u, s, v = la.svd(data_cov)

# 5. 选择特征向量
e = u[:, 0:2]    # 选择两个特征
# 6. 对加载数据降维
de_data = data  @  e
de_data = np.mean(de_data) - de_data    # 中心化还原


from sklearn.decomposition import PCA
pca = PCA(2)
pca.fit(data)
sk_data = pca.transform(data)
# -----  以上代码都是降维

# 7. 可视化(两个坐标系)
figure = plt.figure('我的PCA', figsize=(18, 6))
ax_my = figure.add_subplot(121, title='我的PCA算法')
ax_sk = figure.add_subplot(122, title='sklearn的PCA算法')

# =============================
# SVM对降维数据的分类可视化
# 8. 使用svm进行分类
classfier = sk.svm.SVC(C=5, kernel='rbf', gamma=10)
# 训练
classfier.fit(de_data, target)
# 分类预测
predict = classfier.predict(de_data)
print(predict)

# 9. 可视化分类效果
# 9.1 确定区域范围（取最大最小）
x1_min, x1_max = de_data[:, 0].min(), de_data[:, 0].max()
x2_min, x2_max = de_data[:, 1].min(), de_data[:, 1].max()

x1_min -=0.25
x1_max += 0.25

x2_min -=0.25
x2_max += 0.25

ax_my.set_xlim(x1_min, x1_max)
ax_my.set_ylim(x2_min, x2_max)
# 9.2 把区域形成网格，使用网格生成样本
x1_grid, x2_grid = np.mgrid[x1_min: x1_max:300j, x2_min:x2_max:300j]
samples = np.stack([x1_grid.flat, x2_grid.flat], axis=1)

# 9.3 使用svm预测网格产生的样本
pre = classfier.predict(samples)

# 9.4. 使用网格数据，预测作为颜色，绘制网格区域
bg = mlp.colors.ListedColormap([(0.8, 0.5, 0.5),(0.5, 0.8, 0.8)])
ax_my.pcolormesh(x1_grid, x2_grid,  pre.reshape(x1_grid.shape), cmap=bg)

# =============================

# -----------------------------
# 两个降维方法的结果可视化
colors = mlp.colors.ListedColormap([(0,0,1),(1,0,0)])
ax_my.scatter(de_data[:, 0].flat, de_data[:, 1].flat , c=target, cmap=colors, s=3**2, marker='.')
ax_sk.scatter(sk_data[:, 0].flat, sk_data[:, 1].flat , c=target, cmap=colors, s=3**2, marker='.')


plt.show()
````
