# 感知器与神经网络      

## 感知器与神经网络的基本概念

神经网络来自对人类大脑的模拟，实际是对人脑的一个数学建模。
按照大脑中不同神经元对数据处理的不同功能分类，在神经网络中，包数据处理的基本单位称为神经元（或者称为感知器）

### 感知器的基本要素

1. 输入数据    
2. 数据处理    
3. 输出数据     

###  感知器的分类

根据不同感知处理数据的不同，以及处理方式的不同，可以把感知器分成很多类型，正如人类大脑的神经元分成痛觉神经元、视觉神经元、味觉神经元一样。大致在目前常见的神经元有：           

1. 输入感知器：用于数据输入。     
2. 基本感知器：基本的加权运算，适合大部分数据处理。     
3. 卷积感知器：卷积计算，时候有几何特征的数据处理。       
4. 池化感知器：最值与均值的降维运算。    
5. 方差感知器：方差运算           
6. 循环感知器：适合时序数据的运算        
7. 记忆感知器（长短时记忆感知器）： 适合时序数据的记忆运算           

## 神经网络

感知器（神经元）按照一定的运算结构形成神经网络。按照大脑数据处理的特点，从虚拟模型中构造神经网络也遵循一定的规则。 神经网络遵循如下规则：      

1. 按照相同功能，把感知器分层多层，每层感知器的功能相同；      
2. 下一层感知器只处理上一层感知输出的数据，不处理相同层输出的数据；换句话说就是下一层与上一层相连，相同层之间不相连。   

### 神经网络分层

1. 输入层：用于数据输入处理；      
2. 输出层：数据用于预测与分类处理；      
3. 隐藏层：介于输入层与输出层之间的层，隐藏层用于数据特征处理；           

### 深度神经网络

如果神经网络层数大于等于4层以上，称为深度神经网络（或者隐藏层大于等于2层）      

### 深度神经网络的理解或者优点

深度神经网络与机器学习的最大区别，就是数据的处理形成数据特征的融合与拆分，在融合与拆分过程中，对不重要的特征就丢弃，重要特征就会保留，或者与其他特征融合，或者拆分成多个特征，影响最终的分类与预测作用。

## 感知器的数学模型

已知一个数据样本$X=(x_1，x_2，\dots，x_n)$， 每个特征存在一个加权系数$W=(w_1，w_2，\dots，w_n)$，经过加权求和运算，并对加权求和的结果做一个函数运算$y=f(x)$，得到一个新的输出特征$y$。使用数学公式表示如下：       

$X = \pmatrix { x_1 \\ x_2 \\ \vdots \\ x_n }  \; \;  \to  \sum \limits _{i=1}^{n} w_i x_i = XW  \;\;  \to  y=f(XW) $ 

1. 输入层：每个数据特征就是一个感知器（输入感知器），一共n个感知器。               

2. 激活函数：其中函数$f(x)$根据其作用，称为激活函数 ，为了便于机器学习，其中激活函数需要满足一些条件。    

### 激活函数

1. 激活函数的作用

大家如果还记得逻辑回归，其中在加权求和后，在最后加了一个逻辑分布运算（S函数，sigmoid函数），使得输出的值在$[0,1)$之间，而且从数学上，我们还证明了这个$[0,1)$之间的值可以作为一个概率使用。在感知器最后的函数运算，实际也可以使用逻辑分布函数把输出值变换到$[0,1)$之间，并且把输出值作为概率使用。 但在神经网络之中，激活函数还有其他作用，下面列出激活函数的主要用途：     
    1. 对输出值做处理，用于某些目的，比如使用逻辑分布函数，把输出值限制在$[0,1)$之间，某些输出层可以作为概率使用，也可以防止在多层神经网络训练中，使得值溢出异常等。
    2. 可以通过激活函数，可以决定某些特征在下一层训练中的重要性，比如使用step函数，使得$ \le 0$的输出值，在一层训练中变为0，从而不参与训练。   
    
激活函数的名称来源就是基于第二个用途。

2. 激活函数的数学条件

    【非线性】
        当激活函数是非线性的,然后一个两层神经网络可以证明是一个通用函数近似值。而 identity 激活函数不满足这个属性。当多层网络使用 identity 激活函数，整个网络就相当于一个单层模型.
        
    【连续可微的】
        这个属性对基于梯度优化方法是必要的。二进制激活函数在0点没有可微性，它在其他值上全部可导为0,基于梯度优化方法对于它毫无进展.
        
    【范围】
        当激活函数的范围是有限的，基于梯度的训练方法往往更稳定，因为模式显著影响几个有限权重。当范围是无限的,训练通常更有效率,因为模式显著影响大部分权重，在后一种情况下，较小的学习利率通常是必要的。
        
    【单调】
        当激活函数是单调时，与单层模型相关的错误表面是凸的。
        
    【平滑性】
        有单调导数的平滑函数已经被证明在某些情况下推广地更好。这些属性表明这些激活函数和 “简单有效原理” 更一致.
		
    【原点附近近似identity（f(x)≈x）】
        当激活函数有这个属性，对于小的随机值初始化权重,神经网络将有效地学习，当激活函数没有这个属性,在初始化权值必须使用特殊例子。

3. 常见激活函数

    1. STEP激活函数
    函数定义：
&emsp;&emsp;&emsp;&emsp;$f(x)=\begin{cases} {=1 \;\; x \gt 0} \\ {=0\;\; x \le 0} \end{cases}$      

    导数定义：
&emsp;&emsp;&emsp;&emsp;$f ^{\prime} (x)=0$

    激活函数 Step 更倾向于理论而不是实际，它模仿了生物神经元要么全有要么全无的属性。它无法应用于神经网络，因为其导数是 0（除了零点导数无定义以外），这意味着基于梯度的优化方法并不可行。


    2. Identify激活函数
    函数定义：
&emsp;&emsp;&emsp;&emsp;$f(x)=x$  

    导数定义：
&emsp;&emsp;&emsp;&emsp;$f ^ {\prime} (x)=1$      

    通过激活函数 Identity，节点的输入等于输出。它完美适合于潜在行为是线性（与线性回归相似）的任务。当存在非线性，单独使用该激活函数是不够的，但它依然可以在最终输出节点上作为激活函数用于回归任务。


    3. ReLU激活函数
    函数定义：
&emsp;&emsp;&emsp;&emsp;$f(x)=\begin{cases} {=x \;\; x \gt 0} \\ {=0\;\; x \le 0} \end{cases}$

    导数定义： 
&emsp;&emsp;&emsp;&emsp;$f ^ {\prime}(x)=\begin{cases} {=1 \;\; x \gt 0} \\ {=0\;\; x \le 0} \end{cases}$

    修正线性单元（Rectified linear unit，ReLU）是神经网络中最常用的激活函数。它保留了 step 函数的生物学启发（只有输入超出阈值时神经元才激活），不过当输入为正的时候，导数不为零，从而允许基于梯度的学习（尽管在 x=0 的时候，导数是未定义的）。使用这个函数能使计算变得很快，因为无论是函数还是其导数都不包含复杂的数学运算。然而，当输入为负值的时候，ReLU 的学习速度可能会变得很慢，甚至使神经元直接无效，因为此时输入小于零而梯度为零，从而其权重无法得到更新，在剩下的训练过程中会一直保持静默。


    4. sigmoid激活函数
    函数定义：
&emsp;&emsp;&emsp;&emsp;$S(x)= \dfrac{1}{1 +  e^{-x}}$

    导数定义：
&emsp;&emsp;&emsp;&emsp;$S ^ {\prime}(x)= \dfrac{ e ^{-x} }{{(1 +  e^{-x})}^2}=S(x)(1-S(x))$

    Sigmoid 因其在 logistic 回归中的重要地位而被人熟知，值域在 0 到 1 之间。Logistic Sigmoid（或者按通常的叫法，Sigmoid）激活函数给神经网络引进了概率的概念。它的导数是非零的，并且很容易计算（是其初始输出的函数）。然而，在分类任务中，sigmoid 正逐渐被 Tanh 函数取代作为标准的激活函数，因为后者为奇函数（关于原点对称）。


    5. Tanh激活函数
    函数定义：
&emsp;&emsp;&emsp;&emsp;$f(x) = \dfrac{sinh(x)} {cosh(x) }  = \dfrac{ e^{x} - e^{-x} }{ e^{x} + e^{-x} }$

    导数定义：
&emsp;&emsp;&emsp;&emsp;$f ^ {\prime } (x) = 1 - f ^2 (x)$    
    
    在分类任务中，双曲正切函数（Tanh）逐渐取代 Sigmoid 函数作为标准的激活函数，其具有很多神经网络所钟爱的特征。它是完全可微分的，反对称，对称中心在原点。为了解决学习缓慢和/或梯度消失问题，可以使用这个函数的更加平缓的变体（log-log、softsign、symmetrical sigmoid 等等）

>  **数学科普时间：**     
>  双曲正弦：$sinh(x) = \dfrac { e^{x} - e^{-x} }{ 2 }$      
>
>  双曲余弦：$cosh(x) = \dfrac { e^{x} + e^{-x} }{ 2 }$      
>
>  双曲正切：$tanh(x) = \dfrac{sinh(x)} {cosh(x) }  = \dfrac{ e^{x} - e^{-x} }{ e^{x} + e^{-x} }$      
>
>  双曲余切：$coth(x) = \dfrac{1} {tanh(x) }   = \dfrac{ e^{x} + e^{-x} }{ e^{x} - e^{-x} }$      
>
>  双曲正割：$sech(x) = \dfrac{1} {cosh(x) }  = \dfrac { 2 }{ e^{x} + e^{-x} }$      
>
>  双曲余割：$csch(x) = \dfrac{1} {sinh(x) }   = \dfrac { 2 }{ e^{x} - e^{-x} }$      

    6. loglog激活函数
    函数定义：
&emsp;&emsp;&emsp;&emsp;$f(x) = 1 -  e ^{ e ^ { -x } }$

    导数定义：
&emsp;&emsp;&emsp;&emsp;$f^{ \prime }(x) = e ^ {x} e ^ { e ^ { -x } } = e^{x - e ^ { -x } }$
    
    Log Log 激活函数（由上图 f(x) 可知该函数为以 e 为底的嵌套指数函数）的值域为 [0,1]，Complementary Log Log 激活函数有潜力替代经典的 Sigmoid 激活函数。该函数饱和地更快，且零点值要高于 0.5。

    7. Softsign
    函数定义：
&emsp;&emsp;&emsp;&emsp;$f(x) = \dfrac{ x }{ 1 + | x | } $

    导数定义：
&emsp;&emsp;&emsp;&emsp;$f^{ \prime }(x) = \dfrac{ 1 }{ { ( 1 + | x | ) }^2 } $
    
    Softsign 是 Tanh 激活函数的另一个替代选择。就像 Tanh 一样，Softsign 是反对称、去中心、可微分，并返回-1 和 1 之间的值。其更平坦的曲线与更慢的下降导数表明它可以更高效地学习。另一方面，导数的计算比 Tanh 更麻烦。

    8. Symmetrical Sigmoid
    函数定义：
&emsp;&emsp;&emsp;&emsp;$f(x) = tanh( \dfrac{x}{2}) = \dfrac{ 1- e ^ {- x } } { 1 + e ^ { -x } }$

    导数定义：
&emsp;&emsp;&emsp;&emsp;$f^{ \prime } (x) = \dfrac{1}{2} ( 1 - f ^{2} ( x ) )$
    
    Symmetrical Sigmoid 是另一个 Tanh 激活函数的变种（实际上，它相当于输入减半的 Tanh）。和 Tanh 一样，它是反对称的、零中心、可微分的，值域在 -1 到 1 之间。它更平坦的形状和更慢的下降派生表明它可以更有效地进行学习。

4. 激活函数可视化

```python
# 这里使用Symmetrical Sigmoid函数作为示范例子：  
% matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

figure = plt.figure( figsize=(8, 4) )
ax = figure.add_axes( [0.1, 0.1, 0.8, 0.8] , label='激活函数可视化' )

# 函数的x坐标
x =  np.linspace( -10, 10, 1000 )

# 激活函数的y坐标
y = ( 1 - np.exp( -x ) ) / ( 1 + np.exp( -x ) )

# 导数函数的y坐标
y_ = 1.0 / 2 * ( 1 - y ** 2 ) 

# 绘制激活函数
ax.plot( x, y, color = (1, 0, 0, 1) , linewidth=2, label='激活函数')
# 绘制导数函数
ax.plot( x, y_, color = (0, 0, 1, 1) , linewidth=2, label='导数函数')

plt.legend()
plt.grid(True)
plt.show()
```

### 激活函数对线性性的影响

1. 如果对单层神经网络来说，任何激活函数都影响不了神经网络的线性性，仅仅是多输出值做一个处理。        
2. 如果所有神经网络层都使用Identify激活函数，  不管做多少层神经网络的机器学习，最终的结果也是线性的，只是这个线性是多个线性叠加的结果，最终还是线性的。       
3. 如果在多层神经网络中，每次使用不同的激活函数，最终产生的神经网络可以对非线性数据样本进行分类与预测。     

## 感知器的数学基础与算法     

### 感知器的学习模型 
     
下面的学习模型，指的是基础感知器。      

1. 感知器学习过程

&emsp;&emsp;&emsp;&emsp;（1）随机初始化一个权重矩阵$W$；   
          
&emsp;&emsp;&emsp;&emsp;（2）使用一个数据样本$X_1$，计算加权求和$XW$；    
          
&emsp;&emsp;&emsp;&emsp;（3）使用激活函数对加权求和结果做函数运算$ \bar{y} = f_{activatity} ( XW ) $；     
          
&emsp;&emsp;&emsp;&emsp;（4）对比标签$X_1$的标签$y_i$与输出结果$\bar{y}$，根据误差计算方式计算误差$ \delta $；     
          
&emsp;&emsp;&emsp;&emsp;（5）根据计算出来的误差$ \delta $，计算权重$W$的的调整量$ \nabla $；      
          
&emsp;&emsp;&emsp;&emsp;（6）使用$ \nabla $调整权重矩阵$W$。         
          
&emsp;&emsp;&emsp;&emsp;（7）使用其他样本，反复执行（2）-（6）过程，直到过程（3）计算的误差为0或者可以接受的程度。       
          

2. 感知器学习要素

        （1）激活函数
        （2）误差函数（损失函数）   
        （3）计算提调整量的算法            
        
        （4）学习结果：得到一个【权重矩阵】，使得误差最小（理想结果：损失函数输出为0）      

上述要素是符合我们机器学习的常识的：
        
        （1）当激活函数式Identify函数的时候，感知器就是线性回归。    
        （2）当激活函数是sigmoid函数的时候，感知器就是逻辑回归。      


### 感知器的计算模型

1.  公式一：

&emsp;&emsp;&emsp;&emsp;$ \alpha = XW $
        
2. 公式二：

&emsp;&emsp;&emsp;&emsp;$ \bar{y} = f _ {activatity}(XW) $     

### 感知器的误差模型

在机器学习中，损失函数的通用表达公式如下：    
$\theta ^{*} =  arg \underset{\theta}{min}  \dfrac{1}{N} \sum \limits _{i=1}^ {N}  L (y_i - f(x_i,\theta ))   \; \;  +   \lambda \Phi ( \theta ) $

表示参数$\theta$使得损失函数最小。后面的$\Phi$是关于$\theta$的$L_0,L_1,L_2$范数,用来对损失函数做惩罚项。  （大家可以回顾下线性回归中的岭回归，LASSO回归的损失函数）   

在神经网络中常用损失函数有(实际上神经网络中还定义了很多损失函数)：      

1. 均方差（MSE：Mean Squared Error）   

&emsp;&emsp;&emsp;&emsp;$L = \dfrac{1}{N} \sum \limits _{i=1} ^ {N} {( y_i -\bar{ y_i } )} ^ 2 $      
     
      
&emsp;&emsp;&emsp;&emsp;$y_i$表示标签值，$\bar{y_i}$表示预测输出值

2. sigmoid交叉熵（Sigmoid Cross Entropy）

&emsp;&emsp;&emsp;&emsp;$L=-ln(L(\bar{y_i}))=-\dfrac{1}{N} \sum\limits_{i=1}^{N}y_i\ ln(\bar{y_i}) +(1-y_i)\ln(1-\bar{y_i})$    
     
      
&emsp;&emsp;&emsp;&emsp;$y_i$表示标签值，$\bar{y_i}$表示预测输出值

3. softmax交叉熵（Softamx Cross Entropy）

&emsp;&emsp;&emsp;&emsp;$L = -\dfrac{1}{N} \sum \limits _{i=1} ^ {N}  y_i \;log( \bar{ y_i } )  $      
     
      
&emsp;&emsp;&emsp;&emsp;$y_i$表示标签值，$\bar{y_i}$表示预测输出值

&emsp;&emsp;**注意**：     
        
            决策树中使用的信息熵，本身也是一种损失函数 。

### 感知器的训练模型

1. 计算预测输出值    
&emsp;&emsp;&emsp;&emsp;$\bar{y} = f(XW)$    
2. 计算权重更新梯度

        
&emsp;&emsp;&emsp;&emsp;$\nabla {L}(\omega_i)=L ^ {\prime}(y, \bar{y}) f^\prime(\bar{y} )\ {x_i}$        

        表示第$i$个权重的梯度。     
      
        上述式子可以表示成向量。   


3. 更新权重      
&emsp;&emsp;&emsp;&emsp;$W_{new}  =  W_{old} - \eta \nabla {L}(\omega_i)$     

**提示：**    
        如果上面损失函数采用均方差函数，则导数就是$f ^ {\prime} = \dfrac{2}{N} \sum \limits _{i=1} ^{N} (y_i - \bar{y_i}) $，如果取一个样本就是$f ^ { \prime } =2  (y - \bar{y}) $，当然直接取$f ^ {\prime} =y - \bar{y} $也可以，不影响学习效果， 如果采用所有样本来学习，这就是梯度下降法；如果从样本中随意抽取部分样本（可以抽取一个样本）来学习，这就是随机梯度下降法。     

## 感知器的实现    

### 感知器标量版本实现 

```python
# coding=utf-8
import numpy as np              # 构造权重数组与数学计算
import random                   # 随机数
from sklearn import datasets    # 获取鸢尾花数据样本


class Perceptron_Scalar:
    '''
    感知器的前置条件：
    1.特征数据，特征数据在训练中需要改变，通过函数参数赋值；
    2.特征数据的大小，决定这权重的个数；还要额外设置一个偏置项
    3.激活函数

    注意：这里默认采用随机均方误差损失函数。
    '''
    def __init__(self, input_size, activation_function, activation_derivative, learn_rate):
        '''
        初始化感知器需要的前置条件：
        :param input_size:              输出特征数据的维数大小
        :param activation_function:     激活函数，我们这里使用恒等激活函数f(x)=x
        :param activation_derivative:   激活函数的导数
        :param learn_rate:              学习率
        '''
        # 1.初始化权重向量,大小为输入特征数据大小
        self.weights=np.random.uniform(-0.1, 0.1, input_size)
        self.bias=random.uniform(-0.1, 0.1)
        # 2.激活函数与导数
        self.activation_function = activation_function
        self.activation_derivative = activation_derivative
        # 3.学习率
        self.learn_rate=learn_rate

    def forward(self,input_data):
        '''
        根据已知输入特征数据，进行加权求和与激活函数运算，得到输出。
        :param input_data:              输入的特征数据，必须与构造器指定维数大小一直
        :return:                        返回特征数据计算输出。
        '''
        # 1.加权求和运算
        self.input_data = input_data      # 后面计算更新梯度要使用
        self.sum = 0                      # 加权求和后面要使用
        for idx in range(len(input_data)):
            self.sum += self.weights[idx] * input_data[idx]
        # 偏置项
        self.sum += self.bias
        # 2.激活函数运算
        y = self.activation_function(self.sum)
        self.y = y                        # 后面计算更新梯度要使用
        # print("输出:%f"%self.y)
        return self.y                  # 返回计算输出

    def backward(self, expect_output):
        '''
        :param expect_output:           输入特征数据的期望输出值。
        :return:                        无返回值
        '''
        # 1.计算每个权重更新相同的部分。
        delta = (self.y-expect_output)
        delta = delta*self.activation_derivative(self.sum)
        delta = self.learn_rate * delta   # 学习率

        # 2.计算每个权重对应的更新梯度（包含偏置项）。
        self.w_delta = np.zeros(self.weights.shape, np.float32)
        self.b_delta = 0

        # 偏置项的更新梯度
        self.b_delta=delta
        # 权重的更新梯度
        for idx in range(len(self.w_delta)):
            self.w_delta[idx] = delta*self.input_data[idx]

        # print("权重：{}".format(self.w_delta))

        # 3.更新权重
        self.bias = self.bias-self.b_delta
        for idx in range(len(self.weights)):
            self.weights[idx] = self.weights[idx]- self.w_delta[idx]


class Perceptron_Scalar_App:
    '''
    对感知进行训练，需要训练策略参数：
    1.一个需要被训练的感知器
    '''
    def __init__(self,times,input_size,activation_function,activation_derivative,learn_rate):
        '''
        训练策略参数
        :param times:                   训练迭代次数
        下面参数都是传递给感知器对象的
        :param input_size:
        :param activation_function:
        :param activation_derivative:
        :param learn_rate:
        '''
        self.times=times                # 训练次数
        #
        self.perceptron=Perceptron_Scalar(input_size,activation_function,activation_derivative,learn_rate)

    def train(self,train_data,train_label):
        '''
        训练过程
        :param train_data:              训练的样本数据
        :param train_label:             训练样本的预期输出
        :return:                        无返回值
        '''
        for t in range(self.times):             # 循环训练
            #print("第%04d轮训练"%(t+1))
            for idx in range(len(train_data)):  # 对每个样本数据训练
                # 1.使用感知器对象向前计算结果
                self.perceptron.forward(train_data[idx])
                # 2.使用感知器向后更新权重
                self.perceptron.backward(train_label[idx])

    def predict(self,input_data):
        '''
        对输入数据进行分类：
        :param input_data:              需要分类的特征数据
        :return:                        返回计算结果，结果用于分类
        '''
        return self.perceptron.forward(input_data)      #返回训练后的计算结果
import time
start = time.clock()

#1.加载鸢尾花数据
data,target=datasets.load_iris(return_X_y=True)     #第一个返回值是样本数据，第二个返回值是样本的期望标签
#取前面100个眼本测试（这100个是线性可分的）
train_data=data[:100]
label_data=target[:100]     #第一类是0，第二类是1

#2.构建感知器应用对象
#激活函数
activation_function=lambda x:x
#激活函数导数
activation_derivative=lambda x:1
#感知器应用对象
app=Perceptron_Scalar_App(1000,4,activation_function,activation_derivative,0.001)

#3.训练感知器
app.train(train_data,label_data)

#4.测试感知器(使用训练样本测试)
correct_counter=0
#前50个期望标签都是0（近似0.5以下都算正确）
for item in train_data[:50]:
    y=app.predict(item)
    if y<0.5:
        correct_counter+=1
#后50个期望标签都是1（近似0.5以上都算正确）
for item in train_data[50:]:
    y=app.predict(item)
    if y>=0.5:
        correct_counter+=1

#打印正确率：
print("正确率：%8.2f"%( ( correct_counter / 100.0 ) * 100 ) )

end=time.clock()
print("CPU运行时间：",end-start)
```

### 感知器向量版本实现

```python
# coding=utf-8
import  numpy as np             #构造权重数组与数学计算
import random                   #随机数
from sklearn import datasets    #获取鸢尾花数据样本

class Perceptron_Scalar:
    '''
    感知器的前置条件：
    1.特征数据，特征数据在训练中需要改变，通过函数参数赋值；
    2.特征数据的大小，决定这权重的个数；还要额外设置一个偏置项
    3.激活函数

    注意：这里默认采用随机均方误差损失函数。
    '''
    def __init__(self, input_size, activation_function, activation_derivative, learn_rate):
        '''
        初始化感知器需要的前置条件：
        :param input_size:              输出特征数据的维数大小
        :param activation_function:     激活函数，我们这里使用恒等激活函数f(x)=x
        :param activation_derivative:   激活函数的导数
        :param learn_rate:              学习率
        '''
        # 1.初始化权重向量,大小为输入特征数据大小
        self.weights = np.random.uniform(-0.1,0.1,input_size)
        self.bias = random.uniform(-0.1,0.1)
        # 2.激活函数与导数
        self.activation_function = activation_function
        self.activation_derivative = activation_derivative
        # 3.学习率
        self.learn_rate=learn_rate

    def forward(self, input_data):
        '''
        根据已知输入特征数据，进行加权求和与激活函数运算，得到输出。
        :param input_data:              输入的特征数据，必须与构造器指定维数大小一直
        :return:                        返回特征数据计算输出。
        '''
        # 1.加权求和运算
        self.input_data = input_data      # 后面计算更新梯度要使用
        self.sum = np.dot(self.input_data,self.weights)      # 加权求和后面要使用
        self.sum += self.bias             # 偏置项
        #2.激活函数运算
        y = self.activation_function(self.sum)
        self.y = y                        # 后面计算更新梯度要使用
        #print("输出:%f"%self.y)
        return self.y                  # 返回计算输出

    def backward(self,expect_output):
        '''
        :param expect_output:           输入特征数据的期望输出值。
        :return:                        无返回值
        '''
        # 1.计算每个权重更新相同的部分(delta在这儿是标量)。
        delta = (self.y - expect_output)
        delta = delta * self.activation_derivative(self.sum)
        delta = self.learn_rate * delta   # 学习率

        # 2.计算每个权重对应的更新梯度（包含偏置项）。
        self.w_delta = np.zeros(self.weights.shape, np.float32)    # 向量
        self.b_delta = 0

        # 偏置项的更新梯度
        self.b_delta = delta      # 标量
        # 权重的更新梯度
        self.w_delta = delta*self.input_data  # 向量（哈马达积）

        # 3.更新权重
        self.bias -= self.b_delta
        self.weights -= self.w_delta      # 形状相同的向量与矩阵的减法


class Perceptron_Scalar_App:
    '''
    对感知进行训练，需要训练策略参数：
    1.一个需要被训练的感知器
    '''
    def __init__(self, times, input_size, activation_function, activation_derivative, learn_rate):
        '''
        训练策略参数
        :param times:                   训练迭代次数
        下面参数都是传递给感知器对象的
        :param input_size:
        :param activation_function:
        :param activation_derivative:
        :param learn_rate:
        '''
        self.times = times                #训练次数
        #
        self.perceptron = Perceptron_Scalar(input_size, activation_function, activation_derivative, learn_rate)

    def train(self, train_data, train_label):
        '''
        训练过程
        :param train_data:              训练的样本数据
        :param train_label:             训练样本的预期输出
        :return:                        无返回值
        '''
        for t in range(self.times):             # 循环训练
            # print("第%04d轮训练"%(t+1))
            for idx in range(len(train_data)):  # 对每个样本数据训练
                # 1.使用感知器对象向前计算结果
                self.perceptron.forward(train_data[idx])
                # 2.使用感知器向后更新权重
                self.perceptron.backward(train_label[idx])


    def predict(self,input_data):
        '''
        对输入数据进行分类：
        :param input_data:              需要分类的特征数据
        :return:                        返回计算结果，结果用于分类
        '''
        return self.perceptron.forward(input_data)      # 返回训练后的计算结果


import time
start = time.clock()

# 1.加载鸢尾花数据
data, target=datasets.load_iris(return_X_y=True)     # 第一个返回值是样本数据，第二个返回值是样本的期望标签
# 取前面100个眼本测试（这100个是线性可分的）
train_data = data[:100]
label_data = target[:100]     # 第一类是0，第二类是1

# 2.构建感知器应用对象
# 激活函数
activation_function = lambda x: x
# 激活函数导数
activation_derivative = lambda x: 1
#感知器应用对象
app = Perceptron_Scalar_App(10,4,activation_function,activation_derivative,0.001)

#3.训练感知器
app.train(train_data,label_data)

# 4.测试感知器(使用训练样本测试)
correct_counter=0
# 前50个期望标签都是0（近似0.5以下都算正确）
for item in train_data[:50]:
    y = app.predict(item)
    if y < 0.5:
        correct_counter += 1
# 后50个期望标签都是1（近似0.5以上都算正确）
for item in train_data[50:]:
    y = app.predict(item)
    if y >= 0.5:
        correct_counter += 1

# 打印正确率：
print("正确率：%8.2f"%((correct_counter/100.0)*100))

end = time.clock()
print("CPU运行时间：", end-start)
```

### 感知器Tensorflow版本实现

```python
# coding=utf-8
from sklearn import datasets    #获取鸢尾花数据样本
import tensorflow  as tf        #tensoflow模块
import time

start = time.clock()

#########################################
LEARNING_RATE = 0.001        #学习率
TIMES = 10                    #训练轮数

DATA_SIZE = 4                 #特征个数，也就是训练样本的数组长度

# 一、感知器计算图描述
# 1.描述输入数据：训练样本sample，训练样本的期望标签label
sample = tf.placeholder(dtype=tf.float32, shape=[None, DATA_SIZE])  # 第一个是训练样本个数（不确定设置为None），第二个特征个数
label = tf.placeholder(dtype=tf.float32, shape=[None])                 # 训练样本的期望标签是标量，其值为0或者1

# 2.描述权重与偏置值（初始值随机）
w_init = tf.random_uniform(shape=[DATA_SIZE, 1], minval=-0.1, maxval=0.1, dtype=tf.float32)
weights = tf.Variable(w_init)   # 传统方式

b_init = tf.random_uniform(shape=[], minval=-0.1, maxval=0.1, dtype=tf.float32)
bias = tf.Variable(b_init)

# 3.描述加权求和操作节点
amount = tf.matmul(sample,weights)+bias

# 4.描述损失函数操作节点
loss = tf.reduce_mean(tf.square(amount-label))

#5.描述梯度优化操作节点
trainer=tf.train.GradientDescentOptimizer(LEARNING_RATE).minimize(loss) #使用学习率，最小化误差损失

#二、感知器计算图运行
#加载鸢尾花数据
data,target=datasets.load_iris(return_X_y=True)     #第一个返回值是样本数据，第二个返回值是样本的期望标签
#取前面100个眼本测试（这100个是线性可分的）
train_data=data[:100]
label_data=target[:100]     #第一类是0，第二类是1

#1.构建运行会话
session=tf.Session()
#2.初始化环境
init_op=tf.global_variables_initializer()
session.run(init_op)
#---------------------------
# graph_writer=tf.summary.FileWriter("./graph",graph=session.graph)
#---------------------------
#3.开始训练
print("开始训练！")
for n in range(TIMES):
    # print("第%4d轮训练"%(n+1))
    for i in range(len(train_data)):
        session.run(trainer,feed_dict={sample:train_data[i:i+1],label:label_data[i:i+1]})
    #session.run(trainer, feed_dict={sample: train_data, label:label_data})
print("训练完毕！")

#4.测试训练
correct_counter=0
#前50个期望标签都是0（近似0.5以下都算正确）
for item in train_data[:50]:
    result=session.run(amount,feed_dict={sample:[item]})
    # print(result)
    if result<0.5:
        correct_counter+=1
print("===============")
#后50个期望标签都是1（近似0.5以上都算正确）
for item in train_data[50:]:
    result = session.run(amount, feed_dict={sample: [item]})
    # print(result)
    if result>=0.5:
        correct_counter+=1

#打印正确率：
print("正确率：%8.2f"%((correct_counter/100.0)*100))

#########################################
end=time.clock()
print("CPU运行时间：",end-start)

#-----------------------
# graph_writer.close()
#-----------------------
```

# 单层神经网络

        单层神经网络指的是，只有输入与输出层的神经网络。如果使用鸢尾花作为例子，则输入层是四个输入感知器，输出是两个感知器。这里为什么输出层设计两个感知器呢？  因为我们做二分类问题，一个输出用来表示A类，另外一个输出用来表示B类。 
        
        单层神经网络的数学理论与感知器一样，只是输出是一个向量，表示两个感知器的输出。     
        


        下面直接使用代码说明：

## 单层神经网络矩阵版实现

```python
#coding=utf-8
import numpy as np
from sklearn import  datasets
import time

start = time.clock()

class Layer:
    def __init__(self,input_size,output_size,activity_funtion,derivative_function,learning_rate):
        '''
        初始虎与感知器层有关的数据。
        :param input_size:          输入的特征数据的长度
        :param output_size:         输出的特征数据的长度（就是感知器个数）
        :param activity_funtion:    激活函数
        :param derivative_function: 激活函数的导数
        :param learning_rate:       学习率
        '''
        self.input_size=input_size
        self.output_size=output_size
        #初始化一个随机权重矩阵，矩阵的每行对应一个感知器的权重向量，output_size对应感知器的个数
        self.weights=np.matrix(np.random.uniform(-0.1,0.1,(self.output_size,self.input_size)))
        #加权求和的偏置项(列向量：使用矩阵表示)
        self.bias=np.matrix(np.random.uniform(-0.1,0.1,(self.output_size,1))).T
        #激活函数
        self.activity_funtion=activity_funtion
        #激活函数的导数函数
        self.derivative_function=derivative_function
        #学习率
        self.learning_rate=learning_rate

    def forward(self,input_data):
        '''
        感知器层的输出计算。
        :param input_data:          输入的特征数据
        :return:                    返回感知器的输出
        '''
        self.input_data=input_data
        #计算加权求和（矩阵内积）
        self.weights_sum=self.weights @ self.input_data.T
        #+偏置项（矩阵加法运算）
        self.weights_sum=self.weights_sum + self.bias.T
        #激活函数运算
        self.out_data=self.activity_funtion( self.weights_sum )
        #返回结果(结果是长度为self.output_size的矩阵)
        return self.out_data.T

    def backward(self,delta):
        '''
        根据误差项计算，计算梯度，并使用梯度更新权重
        :param delta:               误差项（根据训练样本的期望标签计算）
        '''
        #计算梯度：梯度=学习率 * 误差项 ● 特征数据
        self.grads = delta.T @ self.input_data
        self.grads *= self.learning_rate
        #更新权重矩阵
        self.weights -= self.grads


class ANN:
    def __init__(self,layers_param):
        '''
        初始化与感知器层有关的参数
        :param layers_param:        所有层感知器相关的参数，使用数组表示一层多个参数
        '''
        self.layers_param=layers_param
        #构造一个感知器层
        self.layer=Layer(layers_param['input_size'],
                         layers_param['output_size'],
                         layers_param['activity_funtion'],
                         layers_param['derivative_function'],
                         layers_param['learning_rate']);

    def forward(self,sample_data):
        '''
        神经网络的输出计算
        :param sample_data:         训练或者测试样本特征数据
        :return:                    返回整个神经网络的输出
        '''
        self.y=self.layer.forward(sample_data)
        return self.y

    def backward(self,label):
        '''
        反向更新权重
        :param label:               与神经网络输入训练样本对应的期望输出标签
        '''
        #根据输出与期望标签，计算误差项（这里损失函数还是采用差平方）。
        #计算输出与期望标签的差
        diff=self.y.T-label.T
        #乘以导数
        delta=diff * self.layers_param['derivative_function'](self.layer.weights_sum)
        #调用感知器层，更新感知器的权重
        self.layer.backward(delta.T)


class ANNApp:
    def __init__(self,ann_param,times):
        '''
        初始化与神经网络的有关的参数
        :param ann_param:           所有神经网络有关的参数
        :param times:               训练次数
        '''
        self.times=times
        #构造一个神经网络
        self.ann=ANN(ann_param)

    def predict(self,sample_data):
        '''
        根据样本数据，计算神经网络输出
        :param sample_data:
        :return:
        '''
        return self.ann.forward(sample_data)

    def train(self,train_dataset,label_dataset):
        #循环训练次数
        for n in range(self.times):
            # print("第%d轮训练"%(n+1))
            #循环训练样本，计算输出，并跟新梯度
            for i in range(len(train_dataset)):
                sample_data=train_dataset[i]
                #计算输出
                self.ann.forward(sample_data)
                #更新权重
                label=label_dataset[i]
                self.ann.backward(label)


PERCEPTRON_SIZE=2
#1.加载数据
data,target=datasets.load_iris(return_X_y=True)
#把期望标签转换为向量
target_num=len(target[:100])
arr_labels=np.zeros((target_num,PERCEPTRON_SIZE))
for i in  range(target_num):
    idx=target[i]
    arr_labels[i][idx]=1

#把数据集转换为矩阵
label_dataset=np.matrix(arr_labels)
train_dataset=np.matrix(data[0:100])
#2.初始化训练相关的参数
times=1000
ann_param={
    'input_size':4,
    'output_size':2,
    'activity_funtion':lambda x:x,
    'derivative_function':lambda x:1,
    'learning_rate':0.001
}
#3.创建神经网络应用对象
app=ANNApp(ann_param,times)
#4.使用训练样本进行神经网络训练
app.train(train_dataset,label_dataset)
#5.使用测试样本测试训练结果
counter=0
for i in range(target_num):
    #计算预测输出
    y=app.predict(train_dataset[i])
    # print(y)
    #判定分类是否正确
    if(y.argmax()==label_dataset[i].argmax()):
        counter+=1

print("准确率：%8.2f"%(100.0*counter/100))
end=time.clock()
print("CPU运行时间：",end-start)
```

# 前馈神经网络


    如果在中间增加几层神经网络层，就得到深度神经网络。
    深度神经网络与单层神经网络基本上模型都一样，但是最大区别就是：误差需要从输出层，反向向前传递到每一层，从而更新每层的权重矩阵。 
    下面使用一幅图可以直观了解多层神经网络。

	1. 多层神经网络（MLP）
		1）MLP（Multi-Layer Perceptron），即多层感知器，是一种前向结构的人工神经网络，映射一组输入向量到一组输出向量。
		2）MLP可以被看做是一个有向图，由多个节点层组成，每一层全连接到下一层。除了输入节点，每个节点都是一个带有非线性激活函数的神经元（或称处理单元）。
		3）一种被称为反向传播算法的监督学习方法常被用来训练MLP。
		4）MLP是感知器的推广，克服了感知器不能对线性不可分数据进行识别的弱点。


	2、激活函数
		1）若每个神经元的激活函数都是线性函数，那么，任意层数的MLP都可被约简化成一个等价的单层感知器。
		2）实际上，MLP本身可以使用任何形式的激活函数，譬如阶梯函数或逻辑乙形函数（logistic sigmoid function），但为了使用反向传播算法进行有效学习，激活函数必须限制为可微函数。
		3）由于具有良好可微性，很多乙形函数，尤其是双曲正切函数（Hyperbolic tangent）及逻辑乙形函数，被采用为激活函数。


	3、MLP应用历史
		1）常被MLP用来进行学习的反向传播算法，在模式识别的领域中算是标准监督学习算法，并在计算神经学及并行分布式处理领域中，持续成为被研究的课题。
		2）MLP已被证明是一种通用的函数近似方法，可以被用来拟合复杂的函数，或解决分类问题。
		3）MLP在80年代的时候曾是相当流行的机器学习方法，拥有广泛的应用场景，譬如语音识别、图像识别、机器翻译等等，但自90年代以来，MLP遇到来自更为简单的支持向量机的强劲竞争。近来，由于深层学习的成功，MLP又重新得到了关注。


## 前馈神经网络模型

### 前馈神经网络向前计算模型

    计算公式：
&emsp;&emsp;&emsp;&emsp;$ y^k = f_k (  y^{k-1} W^k) $，其中$k$表示第$k$层，$f_k$表示第k层的激活函数。每层激活函数可以取成一样。
    
    有的时候，为了更加直观，也会表示成如下方式
&emsp;&emsp;&emsp;&emsp;$ y^k = f_k ( net_{k-1}) $，其中 $net_{k-1}= y^{k-1} W^k$    

### 前馈神经网络向后权重梯度计算模型

        计算公式如下（只考虑一个感知器的情况）：
            输出层梯度计算：
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$ \nabla _ {w_i} =L ^ {\prime}(y, \bar{y}) f^\prime(\bar{y} )\ {x_i} $


&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;可以记: $\delta ^ {k} _ {i}=L ^ {\prime}(y, \bar{y}) f^\prime(\bar{y} )$，称为第$k$层的第$i$个感知器的误差项。


            隐藏层误差项计算：
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\delta ^ {k-1} =f ^ {\prime}(net_{k-1}) \sum \limits _ {i \in K} w^{k} _ {i} \delta ^{k} _{i}   $，其中$K$表示$k$层感知器集合。      
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;实际上第k-1层的误差项可以使用第k层表示。

            隐藏层梯度计算：
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$ \nabla _ {w_i} =\delta ^k {x_i} $


            
            

### 权重更新模型

        公式表示
&emsp;&emsp;&emsp;&emsp;$W_{new}  =  W_{old} - \eta \nabla {L}(W)$     

## 前馈神经网络的数学基础与算法推导 

### 记号   
        
        损失函数：
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$L(W)$

        加权求和与激活函数：
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$a = f(XW） $


        网络层：
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;使用上标表示层，使用下标表示样本的索引。


### 梯度表示

        使用权重系数的偏导数作为梯度
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\nabla {L}(W)= \dfrac { \partial{L(W)} }{ \partial{W} }$

### 梯度的推导

1. 推导的展开思路

&emsp;&emsp;&emsp;&emsp;$L(W) :  L(\bar{y}) \to  L(f) \to L(f, \alpha)  \to L(f, \alpha, W)$    


    不管从哪个部分展开都不是问题，比如从:
&emsp;&emsp;&emsp;&emsp;$f \to \alpha \to  W$

    也可以直接从: 
&emsp;&emsp;&emsp;&emsp;$\alpha \to W$

2. 从$\alpha$开始      
&emsp;&emsp;为了直接剥离输入样本x，所以选择从$\alpha$开始。  

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\nabla {L}(W)= \dfrac { \partial{L(W)} }{ \partial{W} }$

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\nabla {L}(W)= \dfrac { \partial{L(W)} }{ \partial{\alpha} }    \dfrac{\partial{\alpha}}{W}$

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\nabla {L}(W)= \dfrac { \partial{L(W)} }{ \partial{\alpha} }    X$

&emsp;&emsp;为什么这么拆分？ 因为$\alpha$表示一个网络层的核心计算，从形式上讲，只与神经网络每层的输出与最终的损失函数有关，因为每个神经网络结构只与权重系数$W$，激活函数$f$有关，与输入的样本数据没有直接关系。所以我们首先把无关因素剥离出来。

3. $ \dfrac { \partial{L(W)} }{ \partial{\alpha} }$对输出层的推导

 &emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\dfrac { \partial{ L(W) } }{ \partial{\alpha} } = \dfrac { \partial{ L(W) } }{ \partial{f} }    \dfrac{ \partial{f} }  {\alpha}$

 &emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\dfrac { \partial{ L(W) } }{ \partial{\alpha} } =  L^ {\prime}(\bar{y} )    f^ {\prime}(\alpha)$

4. $ \dfrac { \partial{L(W)} }{ \partial{\alpha} }$对隐藏层的推导

 &emsp;&emsp;&emsp;&emsp;其中$\alpha^{k}$表示$k$层，$\alpha^{k-1}$表示$k-1$层，

 &emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\dfrac { \partial{ L(W) } }{ \partial{\alpha}^{k-1} } = \sum \limits _ {i  \in k }  \dfrac { \partial{ L(W) } }{ \partial{\alpha_i}^{k} }    \dfrac{ \partial{ {\alpha_i}^{k}} }{ \partial{\alpha^{k-1}} }$

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\dfrac { \partial{ L(W) } }{ \partial{\alpha}^{k-1} } = \sum \limits _ {i  \in k }  {\delta_i}^{k}    \dfrac{ \partial{ {\alpha_i}^{k}} }{ \partial{\alpha^{k-1}} }$
